# 2007 年 YRBS 青少年酒精使用與體重關聯性研究摘要
# Summary of 2007 YRBS: Alcohol Use and Body Weight Analysis

## 1. 研究問題與動機 (Research Questions)
本研究探討 2007 年美國高中生酒精使用盛行率、平均體重，以及飲酒行為在不同生理性別下對體重與 BMI 的影響。

## 2. 統計推論結果 (Mandatory Statistical Results)

### 3.1 行為變數：比例分析 (Proportion Analysis)
* **研究問題**：2007 年美國高中生飲酒比例是否顯著不同於基準值 0.35？
* **樣本數 (n)**：12,669
* **樣本比例**：45.17%
* **檢定統計量 (Z)**：23.0088
* **P 值 (P-value)**：3.8041e-117
* **結論**：**拒絕虛無假設 (Reject H0)**。證據顯示飲酒盛行率顯著高於 35%。

### 3.2 連續變數：平均數分析 (Mean Analysis)
* **研究問題**：2007 年美國高中生平均體重是否顯著不同於 68.0 kg？
* **樣本數 (n)**：11,843
* **樣本平均數**：68.43 kg
* **檢定統計量 (t)**：2.7429
* **P 值 (P-value)**：6.0993e-03
* **結論**：**拒絕虛無假設 (Reject H0)**。平均體重顯著高於 68.0 公斤。

## 3. 延伸探索：生理性別與酒精之交互作用 (Extended Exploration)
* **分析目標**：探討酒精對體重/BMI 的影響是否因性別而異（更公平的指標）。
* **數據發現**：
    * **男性組別**：飲酒與較高的體重與 BMI 具有顯著正相關（P-value < 0.05）。
    * **女性組別**：飲酒習慣對體重與 BMI 無顯著統計影響（P-value > 0.05）。
* **視覺化證據**：透過交互作用點圖觀察到，男性的平均 BMI 隨飲酒狀態呈現明顯正斜率，而女性則趨於水平。

## 4. 總結 (Conclusion)
研究證實酒精使用對青少年體格指標的影響存在性別調節效應。引入 BMI 作為公平指標後，進一步確認了酒精對男性的影響力高於女性。

# Project Summary: Adolescent Alcohol Use and Body Weight Analysis (2007 YRBS)

## 1. Research Questions & Motivation
This study investigates the prevalence of alcohol consumption and average body weight among U.S. high school students in 2007. Furthermore, it explores the interaction effect between alcohol use and biological sex on physical metrics (Weight and BMI) to provide a more nuanced health analysis.

## 2. Mandatory Statistical Results

### 3.1 Behavioral Variable: Proportion Analysis
* **Research Question**: Is the proportion of students with drinking habits significantly different from the benchmark of 0.35?
* **Sample Size (n)**: 12,669
* **Sample Proportion**: 45.17%
* **Test Statistic (Z)**: 23.0088
* **P-value**: 3.8041e-117
* **Conclusion**: **Reject H0**. There is statistically significant evidence that the drinking rate is higher than 35%.

### 3.2 Continuous Variable: Mean Analysis
* **Research Question**: Is the average body weight of students significantly different from 68.0 kg?
* **Sample Size (n)**: 11,843
* **Sample Mean**: 68.43 kg
* **Test Statistic (t)**: 2.7429
* **P-value**: 6.0993e-03
* **Conclusion**: **Reject H0**. The average body weight is statistically significantly higher than 68.0 kg.

## 3. Extended Exploration: Interaction of Gender and Alcohol
* **Objective**: To determine if the impact of alcohol on body metrics varies by biological sex (using BMI for a fairer comparison).
* **Key Findings**:
    * **Male Group**: Alcohol consumption is significantly associated with higher body weight and BMI (P-value < 0.05).
    * **Female Group**: No statistically significant difference in weight or BMI was found between drinkers and non-drinkers (P-value > 0.05).
* **Visual Evidence**: Interaction point plots show a distinct positive slope for males regarding BMI and alcohol status, while the trend for females remains relatively flat.

## 4. Conclusion
The study confirms a **gender-based moderation effect** regarding the impact of alcohol on adolescent weight metrics. By introducing BMI to control for height variances, the analysis further validates that the physiological or lifestyle impact of alcohol is more pronounced in males than in females within this dataset.
