# 數據重編碼規則 (Recoding Rules)

本專案將原始問卷數據轉換為適合統計推論的格式，具體規則如下：

---

### 1. 酒精使用二元化 (Alcohol Binary Recoding)

為了進行比例分析（Proportion Analysis）與 Z 檢定，我們將原始的飲酒天數變數 `CurrentAlcoholUse` 重新編碼為 `alcohol_binary`：

* **原始編碼與邏輯**：
    * **1** 代表「0 天 (0 days)」。
    * **2~7** 代表「1 天至 30 天」。
* **重編碼規則**：
    * **0 (不喝酒 / Failure)**：原始編碼為 `1`。
    * **1 (有喝酒 / Success)**：原始編碼 `>= 2` (即包含 2, 3, 4, 5, 6, 7)。
* **目的**：建立二元分類，以計算青少年飲酒之盛行率（Success Rate）。

---

### 2. 體重數據處理 (Weight Data Processing)

針對連續型變數 `HowMuchDoYouWeighWithoutShoesInKG`，我們執行了以下處理程序：

* **缺失值移除 (Dropna)**：直接移除所有未填寫體重或填寫無效代碼（如缺失代碼）的樣本。
* **分佈檢查**：
    * 使用**直方圖 (Histogram)** 觀察資料偏態與 KDE 密度。
    * 使用**箱線圖 (Boxplot)** 識別超過 150kg 之極端離群值（Outliers）。
    * 使用**小提琴圖 (Violin Plot)** 綜合評估中位數位置與數據集中趨勢。

---

### 3. 生理性別對照 (Sex Variable)

雖然在 EDA 中維持原始編碼 `1` 與 `2`，但在視覺化標籤中統一對應如下：
* **1** = 女性 (Female)
* **2** = 男性 (Male)
