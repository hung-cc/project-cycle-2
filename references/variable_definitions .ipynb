# 變數定義表 (Variable Definitions)

本專案使用 2007 年青少年危險行為調查 (YRBS) 數據，主要針對酒精使用行為與生理體重進行分析：

| 原始變數名稱 | 變數描述 | 測量方式 / 單位 | 分析樣本數 (n) |
| :--- | :--- | :--- | :--- |
| **CurrentAlcoholUse** | 過去 30 天內的飲酒天數 | 類別變數：1 (0天) 至 7 (30天) | 12,669 (排除缺失後) |
| **HowMuchDoYouWeighWithoutShoesInKG** | 受訪者不穿鞋的體重 | 連續變數：公斤 (kg) | 13,062 (排除缺失後) |
| **alcohol_binary** | 飲酒狀態 (重編碼後) | 二元變數：0 = 不喝酒, 1 = 有喝酒 | 12,669 |

> **數據品質說明**：
> 1. 所有分析均已排除原始數據中的缺失值（NaN）。
> 2. 體重變數已排除非數值之無效代碼，確保分佈統計之準確性。
