# Variable Definitions

This project utilizes data from the 2007 Youth Risk Behavior Surveillance System (YRBSS). The analysis focuses on alcohol consumption behaviors and physiological weight metrics:

| Original Variable Name | Description | Measurement / Units | Sample Size (n) |
| :--- | :--- | :--- | :--- |
| **CurrentAlcoholUse** | Days of alcohol use in the past 30 days | Categorical: 1 (0 days) to 7 (30 days) | 12,669 (Valid cases) |
| **HowMuchDoYouWeighWithoutShoesInKG** | Respondent's weight without shoes | Continuous: Kilograms (kg) | 13,062 (Valid cases) |
| **alcohol_binary** | Alcohol usage status (Recoded) | Binary: 0 = No Use, 1 = Any Use | 12,669 |

> **Data Quality Notes**:
> 1. Missing values (NaN) have been excluded from all statistical computations.
> 2. The weight variable was cleaned by removing invalid entries and specific missing codes (e.g., 979.00) to ensure the integrity of the distribution analysis.
